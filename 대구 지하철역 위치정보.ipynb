```python
import json
import folium
from folium import plugins
```

```python
import pandas as pd
```

```python
xlsx = pd.read_excel("./data/대구도시철도공사_역위치.xlsx")
xlsx.to_csv("./data/대구도시철도공사_역위치.csv")
```

```python
subway = pd.read_csv('./data/대구도시철도공사_역위치.csv', encoding='utf-8')   # 인코딩 오류나면 encoding='cp949 3개중에 하나로 바꿔야됨
subway
```

```python
line2 = subway[subway.선명 == '2호선']
line2.reset_index(drop = True, inplace = True)
line2
```

```python
map = folium.Map(location=[35.864143, 128.437004], tiles='OpenStreetMap',zoom_start = 11)
```

```python
for name,lat, long in zip(line2.역명, line2.위도, line2.경도):
    folium.Marker([lat,long],         # 마커를 표기할 위경도
                  tooltip = name,     # 마우스오버를 했을경우 역 이름이 보이게 설정
                  icon = folium.Icon(prefix = 'fa', icon = 'subway', color = 'green')).add_to(map)
                  # 마커 아이콘 표시
```

```python
plugins.Fullscreen(position='topright',title='Click to Expand',title_cancel='Click to Exit',force_separate_button=True).add_to(map)
plugins.MousePosition().add_to(map)
map
```
